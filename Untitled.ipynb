```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def cal_col(data):
    num_col = data.select_dtypes(include=['float64','int64','int32']).columns
    col = data.columns
    return col,num_col
```

```python
def calculating_iqr(data,column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    q2 = q3 - q1
    upper = q3 + (1.5 * q2)
    lower = q1 - (1.5 * q1)
    return lower,upper
```

```python
handle_outliers = lambda x: lower if x < lower else(upper if x > upper else x)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['temperature','humidity']])
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=50, c='gray', alpha=0.7)
plt.title('Raw Data')
plt.show()
```

```python
# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.3, min_samples=4)
y_dbscan = dbscan.fit_predict(X_scaled)
```

```python
# Visualizing the DBSCAN clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_dbscan, cmap='plasma', s=50)
plt.title('DBSCAN Clustering')
plt.show()
```

```python
df['seasons'] = pd.Series(y_dbscan)
```
